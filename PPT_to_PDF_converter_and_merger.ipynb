# Converter

```python
# Change mode based on what you want: "convert", "merge", "convert_and_merge"
MODE = "convert_and_merge"
INPUT_PATH = "/content/ppt_input"       # folder or single .ppt/.pptx file
OUTPUT_PATH = "/content/pdf_output"     # folder for PDFs / merged file
MERGED_FILENAME = "merged.pdf"
LIBREOFFICE_CMD = "soffice"


def convert_ppt_to_pdf(mode="convert_and_merge"):
    """
    mode: "convert", "merge", "convert_and_merge"
    """
    # ------------------------
    # Convert PPT to PDF
    # ------------------------
    if mode in ["convert", "convert_and_merge"]:
        # Determine files to convert
        if os.path.isfile(INPUT_PATH):
            files = [INPUT_PATH]
        else:
            files = [
                os.path.join(INPUT_PATH, f)
                for f in os.listdir(INPUT_PATH)
                if f.lower().endswith((".ppt", ".pptx"))
            ]

        if not files:
            print("❌ No PPT/PPTX files found to convert.")
            return

        for file in files:
            print(f"➡️ Converting: {file}")
            subprocess.run([
                LIBREOFFICE_CMD,
                "--headless",
                "--convert-to", "pdf",
                "--outdir", OUTPUT_PATH,
                file
            ], check=True)

        print("✅ Conversion complete")

    # ------------------------
    # Merge PDFs
    # ------------------------
    if mode in ["merge", "convert_and_merge"]:
        pdf_files = [
            os.path.join(OUTPUT_PATH, f)
            for f in os.listdir(OUTPUT_PATH)
            if f.lower().endswith(".pdf")
        ]

        if not pdf_files:
            print("❌ No PDFs found to merge.")
            return

        merger = PdfMerger()
        for pdf_file in sorted(pdf_files):
            merger.append(pdf_file)

        merged_path = os.path.join(OUTPUT_PATH, MERGED_FILENAME)
        merger.write(merged_path)
        merger.close()

        print(f"✅ PDFs merged into: {merged_path}")



convert_ppt_to_pdf(MODE)
```

# Downloader

```python
PDF_DIR = "/content/pdf_output"
ZIP_PATH = "/content/all_pdfs.zip"

with zipfile.ZipFile(ZIP_PATH, 'w', zipfile.ZIP_DEFLATED) as zipf:
    for file in os.listdir(PDF_DIR):
        if file.endswith(".pdf"):
            zipf.write(
                os.path.join(PDF_DIR, file),
                arcname=file
            )

print("✅ PDFs zipped successfully")

files.download(ZIP_PATH)
```
